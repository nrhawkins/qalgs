Exercise 4: Equivalent Composition

```python
import qiskit
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
```

```python
qc = QuantumCircuit(3)

qc.ccx(2,1,0)
qc.z(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)
qc.cx(1,0)
qc.h(0)
qc.h(1)
qc.cx(0,2)
qc.cx(2,1)
qc.cx(0,2)
qc.cx(2,1)
qc.h(0)
qc.h(1)
qc.cx(1,0)
qc.h(1)
qc.ccx(0,1,2)

qc.h(1)
qc.h(2)
qc.x(1)
qc.z(2)
qc.h(1)
qc.h(2)
qc.ccx(2,1,0)
qc.ccx(2,1,0)
qc.z(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)
qc.cx(1,0)
qc.h(0)
qc.h(1)
qc.cx(0,2)
qc.cx(2,1)
qc.cx(0,2)
qc.h(0)
qc.cx(2,1)

qc.h(1)
qc.cx(1,0)
qc.h(1)
qc.ccx(0,1,2)
qc.h(1)
qc.h(2)
qc.x(1)
qc.z(2)
qc.h(1)
qc.h(2)
qc.ccx(2,1,0)

qc.draw()
```

```python
original_op = Operator(qc)
print(original_op.data)
```

```python
qcr = QuantumCircuit(3)

qcr.ccx(2,1,0)
qcr.z(1)

qcr.cx(0,2)
qcr.cx(2,1)
qcr.cx(0,2)

qcr.h(0)
qcr.cx(2,1)
qcr.h(1)
qcr.cx(1,0)
qcr.h(1)
qcr.ccx(0,1,2)

qcr.x(2)
qcr.cx(0,2)
qcr.cx(2,1)
qcr.cx(0,2)
qcr.h(0)
qcr.cx(2,1)

qcr.h(1)
qcr.cx(1,0)
qcr.h(1)

qcr.ccx(0,1,2)
qcr.z(1)
qcr.x(2)
qcr.ccx(2,1,0)

qcr.draw()
```

```python
op = Operator(qcr)
print(op.data)
```
